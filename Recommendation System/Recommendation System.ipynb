#### Collaborative Book Recommendation System 

##### Creating Pivot Table

```python
user_books_df = books.pivot_table(index= 'User_ID', columns= 'Book_Title', values= 'Book_Rating').fillna(0)
user_books_df
```

#### Calculating Cosine Similarity Between Users 

```python
ratingsb.sort_values('No_of_Ratings', ascending= False).head(10)
```

```python
stardust_user_ratings = user_books_df['Stardust']
fahrenheit_user_rating = user_books_df['Fahrenheit 451']
fahrenheit_user_rating.head()
```

```python
x_fahrenheit = user_books_df.corrwith(fahrenheit_user_rating)
x_stardust = user_books_df.corrwith(stardust_user_ratings)
```

```python
corr_fahrenheit = pd.DataFrame(x_fahrenheit, columns= ['Correlation'])
corr_fahrenheit.dropna(inplace= True)
corr_fahrenheit.head()
```

```python
corr_fahrenheit.sort_values('Correlation', ascending= False).head(10)
```

```python
corr_fahrenheit = corr_fahrenheit.join(ratingsb['Book_Rating'])
corr_fahrenheit.head()
```

#### Recommending Similar Books based on Ratings


```python
corr_fahrenheit[corr_fahrenheit['Book_Rating']>5].sort_values('Correlation', ascending= False).head()
```

```python
corr_stardust = pd.DataFrame(x_stardust, columns= ['Correlation'])
corr_stardust.dropna(inplace= True)
corr_stardust = corr_stardust.join(ratingsb['No_of_Ratings'])
corr_stardust[corr_stardust['No_of_Ratings']>4].sort_values('Correlation', ascending= False).head()
```

#### Calculating Cosine Similarity Between Users

```python
user_sim = 1 - pairwise_distances(user_books_df.values, metric= 'cosine')
user_sim
```

```python
# Store the results in DataFrame
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```

```python
user_sim_df.iloc[:5,:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5,0:5]
```

```python
# Set the index and column name to User IDs
user_sim_df.index = list(user_books_df.index)
user_sim_df.columns = list(user_books_df.index)
user_sim_df
```

#### cosine similarity recommendation

```python
def give_recommendation(customer_id):
    tem = list(user_sim_df.sort_values([customer_id], ascending= False).head().index)
    books_list = []
    
    for i in tem:
        books_list = books_list + list(books[books['User_ID']==i]['Book_Title'])    
    return set(books_list)-set(books[books['User_ID']==customer_id]['Book_Title'])
```

```python
give_recommendation(14)
```

```python
give_recommendation(8)
```
